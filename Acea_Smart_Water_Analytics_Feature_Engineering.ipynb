Now lets shift the data by combinations of 1 to 31 days and see what gives least error i.e.that number of days shows the real effect of rainfall and temperature

**Depth**

```python
# dataframe for shifted values of depth feature
shifted_depth = pd.DataFrame()
```

```python
#shift the depth values one day at a time
for i in range(1,32):
    shifted_depth['Shifted_Depth_'+str(i)] = aquifer_daily['Mean_Depth'].shift(-i)
```

```python
shifted_depth
```

```python
# fill the gap after shifted values with last value itself
shifted_depth = shifted_depth.ffill()
shifted_depth
```

```python
# take rainfall and temp features
X = aquifer_daily[['Mean_Rainfall','Mean_Temp']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import median_absolute_error
```

```python
# create a random forest model to find the mae between predicted depth and shifted depth
mae_errors = []
for i in range(1,32):
    Y = shifted_depth['Shifted_Depth_'+str(i)]
    X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=12)
    rf = RandomForestRegressor()
    model = rf.fit(X_train, y_train)
    yhat = model.predict(X_test)
    mae = median_absolute_error(y_test,yhat)
    mae_errors.append(mae)
```

```python
# Plot the MAE on depth due to rainfall and temp
plt.title("Median Absolute Error on Depth due to Rain and Temp ")
plt.ylabel("MAE")
plt.xlabel('Day')
plt.plot(mae_errors)
```

**Volume**

```python
# dataframe for shifted values of volume feature
shifted_volume = pd.DataFrame()
for i in range(1,32):
    shifted_volume['Shifted_Vol_'+str(i)] = aquifer_daily['Mean_Volume'].shift(-i)
shifted_volume = shifted_volume.ffill()
shifted_volume
```

```python
# take rainfall and temp features
X = aquifer_daily[['Mean_Rainfall','Mean_Temp']]
```

```python
# create a random forest model to find the mae between predicted depth and shifted volume
mae_errors = []
for i in range(1,32):
    Y = shifted_volume['Shifted_Vol_'+str(i)]
    X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=12)
    rf = RandomForestRegressor()
    model = rf.fit(X_train, y_train)
    yhat = model.predict(X_test)
    mae = median_absolute_error(y_test,yhat)
    mae_errors.append(mae)
```

```python
# Plot the MAE on volume due to rainfall and temp
plt.title("Median Absolute Error on Volume due to Rain and Temp ")
plt.ylabel("MAE")
plt.xlabel('Day')
plt.plot(mae_errors)
```

**Hydrometry**

```python
# dataframe for shifted values of hydrometry feature
shifted_hyd = pd.DataFrame()
for i in range(1,32):
    shifted_hyd['Shifted_Hyd_'+str(i)] = aquifer_daily['Mean_Hydrometry'].shift(-i)
shifted_hyd = shifted_hyd.ffill()
shifted_hyd
```

```python
# create a random forest model to find the mae between predicted depth and shifted hydrometry
X = aquifer_daily[['Mean_Rainfall','Mean_Temp']]
mae_errors = []
for i in range(1,32):
    Y = shifted_hyd['Shifted_Hyd_'+str(i)]
    X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=12)
    rf = RandomForestRegressor()
    model = rf.fit(X_train, y_train)
    yhat = model.predict(X_test)
    mae = median_absolute_error(y_test,yhat)
    mae_errors.append(mae)
```

```python
# Plot the MAE on Hydrometry due to rainfall and temp
plt.title("Median Absolute Error on Hydrometry due to Rain and Temp ")
plt.ylabel("MAE")
plt.xlabel('Day')
plt.plot(mae_errors)
```

We observe that the least error i.e. the actual effect of rainfall and temperature is observed when the values are recorded after 26 days. So we will shift the values by 26 days.

```python
# shift the values in the daily aquifer data shifted to the actual number of days taken to show
# the effect of rainfall and temp
aquifer_daily['Actual_Depth'] = aquifer_daily['Mean_Depth'].shift(-26)
aquifer_daily['Actual_Volume'] = aquifer_daily['Mean_Volume'].shift(-26)
aquifer_daily['Actual_Hydrometry'] = aquifer_daily['Mean_Hydrometry'].shift(-26)
aquifer_daily = aquifer_daily.ffill()
aquifer_daily
```
